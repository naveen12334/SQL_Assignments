# Q2.Write an SQL query to clone a new table from another table.

```python
#sol->
```

```python
cursor.execute('''create table worker_clone([Emp_no.] int,[First_Name] nvarchar(50),[Last_name] nvarchar(50),[Gender] nvarchar(50),[Department] nvarchar(50),[Salary] int,[Date_of_Joining] nvarchar(50),[Email_id] nvarchar(50))''')
conn.commit
```
