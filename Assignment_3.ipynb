# Q-2. Write an SQL query that fetches the unique values of DEPARTMENT from Worker table and prints its length.

```python
#sol->
df = pd.read_sql("SELECT DISTINCT Department, LEN(Department) AS Department_Name_Length FROM worker",conn)
pd.set_option('display.max_columns', None)
df
```

# Q-3. Write an SQL query to fetch nth max salaries from a table.

```python
#sol->
df = pd.read_sql("Select max(Salary) as Salary from worker where Exists (Select TOP (3)[First_Name],[Salary] from worker order by [Salary] desc)",conn)
pd.set_option('display.max_columns', None)
df
```
